```python
# QN: 8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.
#(Makedecision using Hypothesis Testing)
```

```python
#Null Hypothesis: There is no similary between degree_t(Sci&Tech) and specialisation(Mkt&HR) salary
#Alternative Hypothesis: There is similary between degree_t(Sci&Tech) and specialisation(Mkt&HR) salary

#condition: Reject Null hypothesis id p>0.05
```

```python
from scipy.stats import ttest_ind
degree=dataset[dataset["degree_t"]=="Sci&Tech"]["salary"]
spec=dataset[dataset["specialisation"]=="Mkt&HR"]["salary"]
ttest_ind(degree, spec)
```

```python
#Ans: pvalue=0.0078. Since p<0.05, reject null hypothesis and accept alternative hypothesis
```

```python
#10)What is the probability Density Function of the salary range from 700000 to 900000?
```

```python
def pdf(dataset, startrange, endrange):
    import seaborn as sns
    from matplotlib import pyplot
    from scipy.stats import norm
    ax=sns.histplot(dataset, kde=True)
    pyplot.axvline(startrange, color="Red")
    pyplot.axvline(endrange,  color="Red")
    mean=dataset.mean()
    std=dataset.std()
    print("Mean:{} and Std Deviation:{}" .format(mean, std))
    dist=norm(mean, std)
    values = [value for value in range(startrange, endrange)]
    probablities = [dist.pdf(value) for value in values]
    prob=sum(probablities)
    print("The area between the range {}, {} is {}" .format(startrange, endrange, prob))
    return prob
```

```python
pdf(dataset["salary"], 700000, 900000)
```

```python
#Null hypothesis: There is no similarity between degree_t(Sci&Tech) with respect to etest_p and mba_p
#Alternative hypothesis: There is similarity between degree_t(Sci&Tech) with respect to etest_p and mba_p

#condition: Reject null hypothesis if p<0.05
```

```python
from scipy.stats import ttest_rel
degree_t=dataset[dataset["degree_t"] =="Sci&Tech"]['etest_p']
degree_t1=dataset[dataset["degree_t"] =="Sci&Tech"]["mba_p"]
ttest_rel(degree_t, degree_t1)
```

```python
#Answer: pvalue=5.517920600505392e-06. So p>0.05, we are rejecting alternative and accepting null hypothesis 
```
